**Предобработка данных**

Удаление выбросов

```python
#Правило межквантильного расстояния
for p in data_nule:
  q75, q25 = np.percentile(data_nule.loc[:, p], [75, 25])
  iqr = q75 - q25
  max = q75 + (1.5 * iqr)
  min = q25 - (1.5 * iqr)
  data_nule.loc[data_nule[p] < min, p] = np.nan
  data_nule.loc[data_nule[p] > max, p] = np.nan
```

```python
#удаляю нулевые строки
data_clean=data_nule.dropna()
data_clean.info()
```

```python
# Модуль упругости при растяжении
X_ypr=data_norm.drop(['Модуль упругости при растяжении, ГПа'], axis=1)
y_ypr=data_norm['Модуль упругости при растяжении, ГПа']
```

```python
#Разбивка на тестовую и тренировочную выборки
from sklearn.model_selection import train_test_split

X_tr_ypr, X_test_ypr, y_tr_ypr, y_test_ypr = train_test_split(X_ypr, y_ypr, test_size=0.3)
```

```python
#Линейная регрессия
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

model_LRypr = LinearRegression().fit(X_tr_ypr, y_tr_ypr)
y_pred_LRypr = model_LRypr.predict(X_test_ypr)
```

```python
R2_LRypr = model_LRypr.score(X_test_ypr, y_test_ypr)
MSE_LRypr = mean_squared_error(y_test_ypr, y_pred_LRypr)
MAE_LRypr = mean_absolute_error(y_test_ypr, y_pred_LRypr)
print(f'R2: {R2_LRypr}\nMSE: {MSE_LRypr}\nMAE: {MAE_LRypr}')
```

```python
#RandomForestRegressor + CV
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators': [30, 40, 50], 'max_features': [1, 'sqrt', 'log2'], 'min_samples_leaf':[10, 20], 'min_samples_split': [2, 4]} 
model_RFRypr = RandomForestRegressor()
GSCVypr = GridSearchCV(model_RFRypr, param_grid, cv=10)
 
GSCVypr.fit(X_tr_ypr, y_tr_ypr)
```

```python
GSCVypr.best_params_
```

```python
y_pred_RFRypr = GSCVypr.predict(X_test_ypr)
R2_GSCVypr = GSCVypr.score(X_test_ypr, y_test_ypr)
MSE_GSCVypr = mean_squared_error(y_test_ypr, y_pred_RFRypr)
MAE_GSCVypr = mean_absolute_error(y_test_ypr, y_pred_RFRypr)
print(f'R2: {R2_GSCVypr}\nMSE: {MSE_GSCVypr}\nMAE: {MAE_GSCVypr}')
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

```python
#train_dataset = data_clean.sample(random_state=0)
#test_dataset = data_clean.drop(train_dataset.index)

X_NN = data_clean.drop(columns=['Соотношение матрица-наполнитель'])
y_NN = data_clean['Соотношение матрица-наполнитель']
```

```python
#train_features = train_dataset.copy()
#test_features = test_dataset.copy()

#train_labels = train_features.pop('Соотношение матрица-наполнитель')
#test_labels = test_features.pop('Соотношение матрица-наполнитель')

X_NN_train, X_NN_test, y_NN_train, y_NN_test = train_test_split(X_NN, y_NN, test_size=0.3, random_state=42)
```

```python
normalizer = tf.keras.layers.Normalization(axis=-1)
```

```python
normalizer.adapt(np.array(X_NN_train))
```

Линейная модель

```python
linear_model = tf.keras.Sequential([
    normalizer,
    layers.Dense(units=1)
])
```

```python
linear_model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
linear_model.summary()
```

```python
history = linear_model.fit(
    X_NN_train,
    y_NN_train,
    epochs=15,
    verbose=1,
    validation_data=(X_NN_test, y_NN_test))
```

```python
def plot_loss(history):
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.ylim([0, 2])
    plt.xlabel('Epoch')
    plt.ylabel('Error [Соотношение матрица-наполнитель]')
    plt.legend()
    plt.grid(True)
```

```python
plot_loss(history)
```

```python
test_results = {}

test_results['linear_model'] = linear_model.evaluate(
    X_NN_test, y_NN_test, verbose=0)
```

```python
test_predictLM = linear_model.predict(X_NN_test).flatten()

a = plt.axes(aspect='equal')
plt.scatter(y_NN_test, test_predictLM)
plt.title('linear_model')
plt.xlabel('True Values [Соотношение матрица-наполнитель]')
plt.ylabel('Predictions [Соотношение матрица-наполнитель]')
lims = [0, 6]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```

```python
R2_NN_1 = r2_score(y_NN_test, test_predictLM)
R2_NN_1
```

Многослойный персептрон

```python
DNN_model = keras.Sequential([
      normalizer,
      layers.Dense(18, activation='relu'),
      layers.Dropout(0.2),
      layers.Dense(18, activation='relu'),
      layers.Dense(1)
    ])

DNN_model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
```

```python
DNN_model.summary()
```

```python
history = DNN_model.fit(
    X_NN_train,
    y_NN_train,
    epochs=15,
    verbose=1,
    validation_data=(X_NN_test, y_NN_test))
```

```python
plot_loss(history)
```

```python
test_results['DNN_model'] = DNN_model.evaluate(X_NN_test, y_NN_test, verbose=0)
```

```python
test_predictDNN = DNN_model.predict(X_NN_test).flatten()

a = plt.axes(aspect='equal')
plt.scatter(y_NN_test, test_predictDNN)
plt.title('DNN_model')
plt.xlabel('True Values [Соотношение матрица-наполнитель]')
plt.ylabel('Predictions [Соотношение матрица-наполнитель]')
lims = [0, 6]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```

```python
R2_NN_2 = r2_score(y_NN_test, test_predictDNN)
R2_NN_2
```
